# NHTS Data

```python
# from the data dictionary
cols = [
    "year",
    "origin_zone_id",
    "destination_zone_id",
    "annual_total_trips",
    "mode_atf",
    "purpose_work",
    "purpose_nonwork",
    "atf_0_10mi",
    "atf_10_25mi",
    "atf_25_50mi",
    "atf_50_75mi",
    "atf_75_100mi",
    "atf_100_150mi",
    "atf_150_300mi",
    "atf_gt300mi",
    "mode_atf_pct",
    "atf_0_10mi_pct",
    "atf_10_25mi_pct",
    "atf_25_50mi_pct",
    "atf_50_75mi_pct",
    "atf_75_100mi_pct",
    "atf_100_150mi_pct",
    "atf_150_300mi_pct",
    "atf_gt300mi_pct"
]
```

```python
url = "https://nhts.ornl.gov/od/assets/data/National%20Passenger%20Final%20OD%20Data_2020.csv"
%time nhts = pd.read_csv(url, usecols=cols)
```

```python
nhts.info()
```

```python
nhts.head()
```

```python
tot_trips = nhts.groupby("origin_zone_id")[["annual_total_trips", "mode_atf"]].sum()
# create another column, distinct from the mode_atf_p attribute in the original dataset
tot_trips["p_mode_atf"] = tot_trips["mode_atf"] / tot_trips["annual_total_trips"]
tot_trips.reset_index(inplace=True)
```

```python
# total trips originating in the zone - atf
origin_trips = nhts[nhts["origin_zone_id"] != nhts["destination_zone_id"]]
tot_origin_trips = origin_trips.groupby("origin_zone_id")[["annual_total_trips", "mode_atf"]].sum().reset_index()

# total trips ending in the zone - atf
dest_trips = nhts[nhts["destination_zone_id"] != nhts["origin_zone_id"]]
tot_dest_trips = dest_trips.groupby("destination_zone_id")[["annual_total_trips", "mode_atf"]].sum().reset_index()

# total trips within the zone -atf
inzone_trips = nhts[nhts["origin_zone_id"] == nhts["destination_zone_id"]]
tot_inzone_trips = inzone_trips.groupby("origin_zone_id")[["annual_total_trips", "mode_atf"]].sum().reset_index()
```

```python
# merge 
# going to have to rename columns inzone, origin, dest

test = tot_origin_trips.merge(
    tot_dest_trips,
    how="left",
    left_on="origin_zone_id",
    right_on="destination_zone_id",
    suffixes=("_origin", "_dest")
).merge(
    tot_inzone_trips,
    how="left",
    left_on="origin_zone_id",
    right_on="origin_zone_id",
    suffixes=("","_inzone")
)
```
